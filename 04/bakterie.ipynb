```python
# MODEL 1 - implementacja 4 (upływ czasu quasi ciągły)

N = 10 # liczba pokoleń
czas_podwajania = 100 #s
kiedy_podwajac = np.arange(czas_podwajania, N*czas_podwajania, czas_podwajania)
print(kiedy_podwajac)

czas = []
czas.append(0)
populacje = [] # wyniki obliczeń
populacje.append(10) # początkowa ilość bakterii

t = 1 # licznik czasu

while t < N*czas_podwajania:
  
  if (t in kiedy_podwajac):
    populacja_t = 2*populacje[t-1] # Już czas, mnożymy się !
    populacje.append(populacja_t)  #
  else:
    populacja_t = populacje[t-1]  # Nic się nie dzieje, liczebność const
    populacje.append(populacja_t) #

  czas.append(t)
  t = t + 1

print("czas:",czas)
print("populacje w kolejnych pokoleniach:",populacje)

plt.plot(czas, populacje, '.')
plt.xlabel('Czas')
plt.ylabel('Liczebność populacji')

# liczebność "schodki"!
```

```python
# Model 2 logistyczny - założenia
# kolonia identycznych bakterii
# każda komórka dzieli się na dwie po upłynięciu czas podwajania
# komórki żyją wiecznie (nie umierają)
# po osiągnieciu liczebności (gęstości) granicznej, kolonia przestaje się rozmnażać
```

```python
# MODEL 2 - implementacja 3

N = 5 # liczba pokoleń
K = 100 # limit liczebności (przyczyny?) K = 0, 1, 10 , 1000 ...
czas_podwajania = 100 #s
czas = []
czas.append(0)
populacje = [] # wyniki obliczeń
populacje.append(10) # początkowa ilość

print("populacja początkowa: ",populacje)
n = 1
while n < N+1:
  populacja_n = populacje[n-1] + (1 - populacje[n-1]/K) * populacje[n-1] # Równanie logistycznego wzrostu !
  populacje.append(populacja_n)
  czas.append(czas[n-1]+czas_podwajania)
  n = n + 1

print("czas:",czas)
print("populacje w kolejnych pokoleniach:",populacje)

plt.plot(czas, populacje, 'o-')
plt.xlabel('Czas')
plt.ylabel('Liczebność populacji')

# porównać z Model 1 implementacja 3
# wykres w układzie log-log
```

```python
# GRA w życie komórek
#
#


# adaptacja ####
mu_lag = 20
sigma_lag = 3
# podziały komórkowe ####
mu_div = 100
sigma_div = 3
# śmierć ####
mu_death = 400
sigma_death = 40




# statystyka czasu adaptacji

nums = []

for i in range(10000):
  temp = rnd.gauss(mu_lag, sigma_lag)
  nums.append(temp)
  
# plotting a graph 
plt.title('Statystyka osiągnięcia czasu adaptacji')
plt.hist(nums, bins = 200, color='c')
plt.xlabel('Czas adaptacji')
plt.ylabel('Prawdopodobieństwo zdarzenia')
plt.show()

# statystyka czasu podziału komórkowego

nums = []
for i in range(10000):
  temp = rnd.gauss(mu_div, sigma_div)
  nums.append(temp)
  
# plotting a graph 
plt.title('Statystyka podziału komórki')
plt.hist(nums, bins = 200, color='g')
plt.xlabel('Czas')
plt.ylabel('Prawdopodobieństwo zdarzenia')
plt.show()

# statystyka czasu śmierci

nums = []
for i in range(10000):
  temp = rnd.gauss(mu_death, sigma_death)
  nums.append(temp)
  
plt.title('Statystyka śmierci komórki')
plt.hist(nums, bins = 200, color='k')
plt.xlabel('Czas życia')
plt.ylabel('Prawdopodobieństwo zdarzenia')
plt.show()
```

```python

NG = 100 # liczba pokoleń #########

S_LAG = 0
S_GROWTH = 1
S_DEATH = 2


class Cell:
  def __init__(self):
    self.age = 0
    self.state = S_LAG

    # w momencie urodzin komórki LOSOWANE (gauss) są momenty podziału i czas śmierci komórki
    self.lag_time = int(rnd.gauss(mu_lag, sigma_lag))
    self.div_time = [int(rnd.gauss(mu_div*i, sigma_div)) for i in range(1,NG+1)]
    self.death_time = int(rnd.gauss(mu_death, sigma_death))
  
  def info(self):
    print("age:", self.age)
    print("state:", self.state)
    print("lag:", self.lag_time)
    print("div:", self.div_time)
    print("death:", self.death_time)

```

```python
czas_symulacji = 1000
N0 = 10 ###

medium = []
for i in range(N0):
  medium.append(Cell())

print(medium)

# przykładowo, inspekcja komórki nr 4
medium[4].info()

time = []
population = []

for t in range(czas_symulacji):
  
  #print(t, len(medium))
  c = 0

  to_kill = []
  to_born = 0

  for cell in medium:
    
    cell.age = cell.age + 1
    
    if (cell.state == S_LAG): # lag faza
      if(cell.age == cell.lag_time):
        cell.state = S_GROWTH # zakończenie lag wejście we wzrost

    else: # wzrost lub umieranie
      if(cell.age == cell.death_time):
        to_kill.append(c)
        
      elif(cell.age in cell.div_time):
        r = rnd.random()
        if(r>0.5):
          to_born = to_born + 1

    c = c + 1

  for c in to_kill:
    del medium[c]
    print(c)
  for i in range(to_born+1):
    medium.append(Cell())

  time.append(t)
  population.append(len(medium))
 

        


```
